```python
# Import libraries
import pandas as pd
```

```python
# Import data
path = "../Data/UTF_Data3.csv"
df = pd.read_csv(path, sep=';')
df.head(5)
```

```python
def removeRareColumns(df, threshold=0.91):
    dfLength = len(df)
    for column in df:
        count = df[column].count()
        percent = count / dfLength

        if percent < threshold:
            print(column)
            df = df.drop(column, axis=1)
        else:
            pass
    return df

print("Columns before: ", len(df.columns))
df = removeRareColumns(df)
print("Columns after: ", len(df.columns))

# Remove NaNs
print("Rows before: ", len(df))
df = df.dropna()
print("Rows after: ", len(df))
df.head()
```
