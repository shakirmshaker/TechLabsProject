```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import datetime as dt
from sklearn.linear_model import LinearRegression 
import pprint
```

```python
df = pd.read_csv('Data/DatasetWithDateInfo', sep=',')
df.head()
```

```python
df1 = pd.read_csv('Data/DatasetWithDateInfo.csv', sep=',')
df1.head()
```

```python
df2 = pd.read_csv('Data/Merged_Dataset.csv', sep=';')
df2.head()
```

```python
df1['Date'] = df2['WDate']
df1.head()
```

### Dataframe for plotting

```python
# Create new dataframe for plotting
df3 = df1.copy()

# Move predictor variable to the end for convenience, and drop unnecessary columns
predictor = df3.TotalCount.copy()

df3 = df3.drop(columns=['Unnamed: 0', 'TotalCount'])
df3['TotalCount'] = predictor
df3.head(3)
```

#### Create inverals for HpA

```python
# Define min, max, and interval count
#df3.HpA.describe()
HpA_min = int(df3.HpA.min())
HpA_max = int(df3.HpA.max())
HpA_interval = 20

# Create interval
intervalsList = list(range(HpA_min,HpA_max,HpA_interval))
bins = pd.cut(x=df3.HpA, bins=HpA_interval)
df3.insert(2, 'HpA_bins', bins)
df3.head(3)
```

#### Create averages

```python
# Create empty dictionary for storage of average values
dict_Averages = {}

# Define columns
columns = df3.columns

# Calculate average values
for column in columns[:-1]:
    print('Creating averages for: {}'.format(column))
    
    # Unique values in column
    col_unique = df3[column].unique()
    
    # Add empry column to dict
    dict_Averages[column] = {}
    
    # Find averages
    for value in col_unique:
        avg = df3[df3[column] == value].TotalCount.mean()
        
        # add to dict
        dict_Averages[column][value] = avg
        
# Print results
print('Printing averages:')
pprint.pprint(dict_Averages)
```

### Plotting averages

```python
# Pyplot: Degrees
# Unpack averages from dict
lists = sorted(dict_Averages['Degrees_C'].items()) # sorted by key, return a list of tuples
degrees, count_degree = zip(*lists) # unpack a list of pairs into two tuples
   
fig = plt.figure(figsize = (10, 5)) 
  
# creating the bar plot 
plt.bar(degrees, count_degree, color ='c',  
        width = 0.5) 
  
plt.xlabel("Temperature") 
plt.ylabel("Avg. count") 
plt.title("Title") 
plt.show()
```

```python
# Pyplot: HpA
# Unpack averages from dict
lists = dict_Averages['HpA_bins'].items() # sorted by key, return a list of tuples
HpA, count_HpA = zip(*lists) # unpack a list of pairs into two tuples
HpA = list(HpA)[:-1]
count_HpA = list(count_HpA)[:-1]
for i, interval in enumerate(HpA):
    left = interval.left
    right = interval.right
    HpA[i] = "[{},{}]".format(left,right)
    
fig = plt.figure(figsize = (44, 5)) 
  
# creating the bar plot 
plt.bar(HpA, count_HpA, color ='c',  
        width = 0.5) 
  
plt.xlabel("HpA") 
plt.ylabel("Avg. count") 
plt.title("Title") 
plt.show()
```

```python
# Pyplot: Humidity
# Unpack averages from dict
lists = sorted(dict_Averages['Humidity'].items()) # sorted by key, return a list of tuples
Humidity, count_hum = zip(*lists) # unpack a list of pairs into two tuples
   
fig = plt.figure(figsize = (10, 5)) 
  
# creating the bar plot 
plt.bar(Humidity, count_hum, color ='c',  
        width = 0.5) 
  
plt.xlabel("Humidity") 
plt.ylabel("Avg. count") 
plt.title("Title") 
plt.show()
```

```python
# Pyplot: Wind_Speed
# Unpack averages from dict
lists = sorted(dict_Averages['Wind_Speed'].items()) # sorted by key, return a list of tuples
Wind_Speed, count_WS = zip(*lists) # unpack a list of pairs into two tuples
   
fig = plt.figure(figsize = (10, 5)) 
  
# creating the bar plot 
plt.bar(Wind_Speed, count_WS, color ='c',  
        width = 0.5) 
  
plt.xlabel("Wind Speed") 
plt.ylabel("Avg. count") 
plt.title("Title") 
plt.show()
```

```python
# Create subplots
fig = plt.figure(figsize = (20, 10))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.bar(degrees, count_degree, color ='c',  
        width = 0.5)
ax2.bar(HpA, count_HpA, color ='c',  
        width = 0.5)
ax3.bar(Humidity, count_hum, color ='c',  
        width = 0.5)
ax4.bar(Wind_Speed, count_WS, color ='c',  
        width = 0.5)

ax1.title.set_text('Temperature')
ax2.title.set_text('HpA')
ax3.title.set_text('Humidity')
ax4.title.set_text('Wind Speed')

ax1.set_ylabel('Avg. count')
ax2.set_ylabel('Avg. count')
ax3.set_ylabel('Avg. count')
ax4.set_ylabel('Avg. count')

plt.show()
```
